```python
alphabet = ['a','b','c','d','e','f','g','h','i','i','j','k','l','m','n',
            'o','p','q','r','s','t','u','v','w','y','z']



def encrypt(original_text,shift_amount):
    encrypted_text = ""
    for letter in original_text.lower():
        encrypted_text += alphabet[(alphabet.index(letter) + shift_amount)%len(alphabet)]
    return encrypted_text

def decrypt(encrypted_text,shift_amount):
    decrypted_text = ""
    for letter in encrypted_text.lower():
        decrypted_text += alphabet[(alphabet.index(letter) - shift_amount + 26)%len(alphabet)]
    return decrypted_text

def ceasar(original_text,shift_amount, encode_or_decode):
    """Encrypt/decrypt a word using the shift amount provided"""
    output_text = ""
    if encode_or_decode == "decode":
            shift_amount *= -1
    for letter in original_text.lower():
        output_text += alphabet[(alphabet.index(letter) + shift_amount+ 26)%len(alphabet)]
    return output_text


print(encrypt("zebra",3))
print(decrypt("cheud",3))
print(ceasar("zebra",3,"encode"))
print(ceasar("cheud",3,"decode"))
```

# Blackjack Game Project

```python
cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]

import random
import time


def draw_card():
    return random.choice(cards)


def initial_draw():
    drawn_cards = []
    for card in range(0, 2):
        drawn_cards.append(draw_card())
    return drawn_cards


player_cards = initial_draw()
dealer_cards = initial_draw()


def sum_cards(card_list):
    if sum(card_list) <= 21:
        return sum(card_list)
    elif sum(card_list) > 21 and 11 in card_list:
        card_list.remove(11)
        card_list.append(1)
        return sum(card_list)
    else:
        return sum(card_list)

game_over = False
while sum_cards(player_cards) < 21 and sum_cards(dealer_cards) < 21 and game_over == False:

    print(f"Your cards are {player_cards}. Sum = {sum_cards(player_cards)}")
    print(f"Dealer cards are {dealer_cards[0]} and x")

    if sum_cards(player_cards) == 21:
        print("You win!")
        game_over = True
        break
    elif sum_cards(player_cards) > 21:
        print("You lose!")
        game_over = True
        break
    else:
        while sum_cards(dealer_cards) < 21 and sum_cards(dealer_cards) < 21 and game_over == False:
            user_choice = input("Do you want to hit another card? Y/N\t")

            if user_choice.lower() == "y":
                player_cards.append(draw_card())
                print(f"Your cards are {player_cards}. Sum = {sum_cards(player_cards)}")
                if sum_cards(player_cards) == 21:
                    print("You win!")
                    game_over = True
                    break
                elif sum_cards(player_cards) > 21:
                    print("You lose!")
                    game_over = True
                    break

            elif user_choice.lower() == "n":
                print(f"Dealer cards are {dealer_cards}. Sum = {sum_cards(dealer_cards)}")
                while sum_cards(dealer_cards) < 17:
                    dealer_cards.append(draw_card())
                    print("Dealer takes another card")
                    print(f"Dealer cards are {dealer_cards}. Sum = {sum_cards(dealer_cards)}")
                    if 21 >= sum_cards(dealer_cards) > sum_cards(player_cards):
                        print("You lose!")
                        game_over = True
                        break
                    elif sum_cards(dealer_cards) > 21:
                        print("You win!")
                        game_over = True
                        break
                    time.sleep(3)
                if sum_cards(dealer_cards) == sum_cards(player_cards):
                    print("It's a draw!")
                    game_over = True
                    break

                elif sum_cards(dealer_cards) < sum_cards(player_cards):
                    print("You win!")
                    game_over = True
                    break





```

# Namespace - Global vs Local Scope

```python
 # Functions count as local scope. Blocks like if, for, while loops doesn't count as local scope

 a = 1

 def increase_value():
     a = 2
     print(f"Value inside function is {a}")      # Local scope

increase_value()
print(f"Value outside function is {a}")     # Global scope
```

```python
 # Functions count as local scope. Blocks like if, for, while loops doesn't count as local scope

 a = 1

 def increase_value():
     global a            # Global variable called locally
     a = 2
     print(f"Value inside function is {a}")

increase_value()
print(f"Value outside function is {a}")     # Global scope
```

```python
# Best practice

a = 1

def increase_value(a):
    a +=1
    print(f"Value inside function is {a}")
    return a

a = increase_value(a)
print(f"Value outside function is {a}")     # Global scope
```

```python
class User:
    def __init__(self, id, name):
        # initialise attributes
        self.id = id
        self.name = name
        self.followers = 0
        self.followers_list = []

    def follow_user(self,follow_list):
        self.followers += len(follow_list)
        self.followers_list.extend(follow_list)

    def unfollow_user(self,unfollow_list):
        self.followers -= len(unfollow_list)
        # for user in unfollow_list:
        #     if user in self.followers_list:
        #         self.followers_list.remove(user)
        self.followers_list = [user for user in self.followers_list if user not in unfollow_list]             # List comprehension

user1 = User(id="001", name="Deepak")
user2 = User(id="002", name="John")

user1.follow_user(['John', 'Orla','Gen'])
user1.follow_user(['Aaron','Mart'])
user1.unfollow_user(['Orla'])
user1.follow_user(['Akash','Jeff'])

print(user1)
print(user1.id)
print(user1.name)
print(user1.followers)
print(user1.followers_list)
```

```python
class User:
    def __init__(self, id, name):
        # initialise attributes
        self.id = id
        self.name = name
        self.followers = 0
        self.following = 0

    def follow(self,user):
        self.following += 1
        user.followers += 1

user1 = User(id="001", name="Deepak")
user2 = User(id="002", name="John")
user3 = User(id="003", name="Jeff")

user1.follow(user2)
user1.follow(user3)
user3.follow(user1)


print(user1.followers)
print(user1.following)
print(user2.followers)
print(user2.following)
print(user3.followers)
print(user3.following)
```

```python
class Car:
    def __init__(self, name, seats, color):
        self.name = name                      # Attributes
        self.seats = seats
        self.color = color

    def enter_race_mode(self):               # Method
        self.seats = 2


car1 = Car(name = 'Polo',seats=5, color="red")
car2 = Car(name = 'Audi',seats=4, color="blue")

car1.enter_race_mode()
print(car1)
print(car1.seats)
```

Class Inheritance

```python
class SportsCar(Car):
    def __init__(self,name,seats,color,max_speed):
        super().__init__(name,seats,color)
        self.max_speed = max_speed
sports_car = SportsCar("Ferrari",2,"Red",450)
print(sports_car)
print(sports_car.max_speed)
```
