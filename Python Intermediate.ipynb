# List Comprehension

```python
a = [1,2,3,4,5]
print(a)

b = [i**2 for i in a]
print(b)
```

```python
word = 'Batman'
letters = [letter for letter in word]
print(letters)
```

```python
# Conditional list comprehension

names = ['Ben','Beth','Carl','Denise','Mark','Hilton','James']
print(names)

short_names = [name for name in names if len(name) <= 4]
print(short_names)
```

```python
# Dictionary comprehension
import random
names = ['Ben','Beth','Carl','Denise','Mark','Hilton','James']
print(names)

names_dic = {name:len(name) for name in names}
print(names_dic)

student_scores = {name:random.randint(0,100) for name in names}
print(student_scores)

new_names_dic = {key:key.upper() for (key,value) in names_dic.items()}
print(new_names_dic)

passed_students = {name:score for (name,score) in student_scores.items() if score> 60}
print(passed_students)
```
